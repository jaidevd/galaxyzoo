##### Loading and preprocessing the data.

```python
# imports
import os
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
```

```python
filename = "/Users/jaidevd/GitHub/kaggle/galaxy/solutions_training.csv"
solutions = pd.read_csv(filename, index_col=0)
data = solutions.values
```

###### Simple PCA in 2D and 3D of the solution data

```python
# 2D PCA
pca2d = PCA(n_components=2)
x_red2 = pca2d.fit_transform(data)
figure(figsize=(10,10))
plot(x_red2[:,0],x_red2[:,1],'b^')
```

```python
# 3D PCA
from mpl_toolkits.mplot3d import Axes3D
pca3d = PCA(n_components=3)
x_red3 = pca3d.fit_transform(data)
fig = figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_red3[:,0], x_red3[:,1], x_red3[:,2], 'b^')
```

###### KMeans clustering of principal components

```python
mbkm = MiniBatchKMeans(n_clusters=8)
mbkm.fit(x_red3)
```

```python
# 3D projection
centers3 = mbkm.cluster_centers_
fig = figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
# Plot the centers
ax.scatter(centers3[:,0], centers3[:,1], centers3[:,2], marker='^', c='r', s=100)
```

```python
# 2D projection
mbkm2 = MiniBatchKMeans(n_clusters=10)
mbkm2.fit(x_red2)
```

```python
centers2 = mbkm2.cluster_centers_
figure(figsize=(10,10))
plot(x_red2[:,0], x_red2[:,1], 'go', markersize=5)
plot(centers2[:,0], centers2[:,1], 'r^', markersize=10)
```
